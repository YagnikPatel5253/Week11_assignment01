Q1. What is the Probability density function?

```python
import numpy as np
import matplotlib.pyplot as plt

# Define parameters of the normal distribution
mu = 0  # Mean
sigma = 1  # Standard deviation

# Define the probability density function (PDF) for a normal distribution
def normal_pdf(x, mu, sigma):
    return 1 / (sigma * np.sqrt(2 * np.pi)) * np.exp(-0.5 * ((x - mu) / sigma)**2)

# Generate x values
x_values = np.linspace(-5, 5, 1000)

# Compute y values using the PDF
y_values = normal_pdf(x_values, mu, sigma)

# Plot the PDF
plt.figure(figsize=(8, 6))
plt.plot(x_values, y_values, label='Normal PDF (mu=0, sigma=1)')
plt.title('Probability Density Function (PDF) of a Normal Distribution')
plt.xlabel('x')
plt.ylabel('Probability Density')
plt.legend()
plt.grid(True)
plt.show()
```

Q2. What are the types of Probability distribution?

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import bernoulli, binom, poisson, norm, uniform, expon, gamma

# Set up the figure and axis
fig, axs = plt.subplots(3, 2, figsize=(12, 12))

# Example 1: Bernoulli Distribution
p = 0.3  # Probability of success
x = [0, 1]
axs[0, 0].bar(x, bernoulli.pmf(x, p), color='blue')
axs[0, 0].set_title('Bernoulli Distribution (p=0.3)')
axs[0, 0].set_xticks(x)
axs[0, 0].set_xlabel('X')
axs[0, 0].set_ylabel('Probability')

# Example 2: Binomial Distribution
n = 10  # Number of trials
p = 0.3  # Probability of success
x = np.arange(0, n+1)
axs[0, 1].bar(x, binom.pmf(x, n, p), color='green')
axs[0, 1].set_title('Binomial Distribution (n=10, p=0.3)')
axs[0, 1].set_xticks(x)
axs[0, 1].set_xlabel('X')
axs[0, 1].set_ylabel('Probability')

# Example 3: Poisson Distribution
mu = 3  # Average rate
x = np.arange(0, 10)
axs[1, 0].bar(x, poisson.pmf(x, mu), color='red')
axs[1, 0].set_title('Poisson Distribution (mu=3)')
axs[1, 0].set_xticks(x)
axs[1, 0].set_xlabel('X')
axs[1, 0].set_ylabel('Probability')

# Example 4: Normal Distribution
mu = 0
sigma = 1
x = np.linspace(-5, 5, 100)
axs[1, 1].plot(x, norm.pdf(x, mu, sigma), color='purple')
axs[1, 1].set_title('Normal Distribution (mu=0, sigma=1)')
axs[1, 1].set_xlabel('X')
axs[1, 1].set_ylabel('Probability Density')

# Example 5: Uniform Distribution
a = 0
b = 1
x = np.linspace(a, b, 100)
axs[2, 0].plot(x, uniform.pdf(x, a, b-a), color='orange')
axs[2, 0].set_title('Uniform Distribution (a=0, b=1)')
axs[2, 0].set_xlabel('X')
axs[2, 0].set_ylabel('Probability Density')

# Example 6: Exponential Distribution
lambda_ = 0.5
x = np.linspace(0, 5, 100)
axs[2, 1].plot(x, expon.pdf(x, scale=1/lambda_), color='brown')
axs[2, 1].set_title('Exponential Distribution (lambda=0.5)')
axs[2, 1].set_xlabel('X')
axs[2, 1].set_ylabel('Probability Density')

plt.tight_layout()
plt.show()
```

Q4. What are the properties of Binomial distribution? Give two examples of events where binomial
distribution can be applied.

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import binom

# Example 1: Coin Flipping
n_flips = 10  # Number of coin flips
p_heads = 0.5  # Probability of getting heads (success)
x_flips = np.arange(0, n_flips+1)
plt.bar(x_flips, binom.pmf(x_flips, n_flips, p_heads), color='blue')
plt.title('Binomial Distribution (Coin Flipping)')
plt.xlabel('Number of Heads')
plt.ylabel('Probability')
plt.xticks(x_flips)
plt.grid(True)
plt.show()

# Example 2: Manufacturing Quality Control
n_items = 20  # Number of items produced
p_defective = 0.1  # Probability of an item being defective
x_defective = np.arange(0, n_items+1)
plt.bar(x_defective, binom.pmf(x_defective, n_items, p_defective), color='green')
plt.title('Binomial Distribution (Manufacturing Quality Control)')
plt.xlabel('Number of Defective Items')
plt.ylabel('Probability')
plt.xticks(x_defective)
plt.grid(True)
plt.show()
```

Q6. Write a Python function to calculate the cumulative distribution function of a Poisson distribution
with given mean at a given point.

```python
from math import exp
from scipy.special import gamma

def poisson_cdf(x, lambd):
    cdf = 0
    for k in range(int(x) + 1):
        cdf += (exp(-lambd) * (lambd ** k)) / gamma(k + 1)
    return cdf

# Example usage:
mean = 2.5
point = 3
cdf_value = poisson_cdf(point, mean)
print("CDF at x =", point, ":", cdf_value)
```

Q7. How Binomial distribution different from Poisson distribution?

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import binom, poisson

# Parameters
n_trials = 1000  # Number of trials or observations
p_success = 0.3  # Probability of success for Binomial distribution
lambda_poisson = 300  # Average rate for Poisson distribution

# Generate random samples
binomial_sample = np.random.binomial(n_trials, p_success, size=1000)
poisson_sample = np.random.poisson(lambda_poisson, size=1000)

# Plot histograms
plt.figure(figsize=(12, 6))

plt.subplot(1, 2, 1)
plt.hist(binomial_sample, bins=range(0, n_trials+1), density=True, color='skyblue', alpha=0.7, label='Binomial')
plt.title('Binomial Distribution Example')
plt.xlabel('Number of Successes')
plt.ylabel('Probability')
plt.legend()
plt.grid(True)

plt.subplot(1, 2, 2)
plt.hist(poisson_sample, bins=range(0, max(poisson_sample)+1), density=True, color='lightgreen', alpha=0.7, label='Poisson')
plt.title('Poisson Distribution Example')
plt.xlabel('Number of Events')
plt.ylabel('Probability')
plt.legend()
plt.grid(True)

plt.tight_layout()
plt.show()
```

Q8. Generate a random sample of size 1000 from a Poisson distribution with mean 5 and calculate the
sample mean and variance.

```python
import numpy as np

# Parameters
mean = 5
sample_size = 1000

# Generate random sample from Poisson distribution
random_sample = np.random.poisson(mean, size=sample_size)

# Calculate sample mean and variance
sample_mean = np.mean(random_sample)
sample_variance = np.var(random_sample)

print("Sample Mean:", sample_mean)
print("Sample Variance:", sample_variance)
```

Q9. How mean and variance are related in Binomial distribution and Poisson distribution?

```python
import numpy as np
import matplotlib.pyplot as plt

# Parameters
n_trials = 1000  # Number of trials or observations
p_success_binomial = 0.3  # Probability of success for Binomial distribution
lambda_poisson = 300  # Average rate for Poisson distribution

# Generate random samples
binomial_sample = np.random.binomial(n_trials, p_success_binomial, size=1000)
poisson_sample = np.random.poisson(lambda_poisson, size=1000)

# Calculate mean and variance
binomial_mean = np.mean(binomial_sample)
binomial_variance = np.var(binomial_sample)
poisson_mean = np.mean(poisson_sample)
poisson_variance = np.var(poisson_sample)

# Plot histograms
plt.figure(figsize=(12, 6))

plt.subplot(1, 2, 1)
plt.hist(binomial_sample, bins=range(0, n_trials+1), density=True, color='skyblue', alpha=0.7, label='Binomial')
plt.title('Binomial Distribution (Mean={:.2f}, Variance={:.2f})'.format(binomial_mean, binomial_variance))
plt.xlabel('Number of Successes')
plt.ylabel('Probability')
plt.legend()
plt.grid(True)

plt.subplot(1, 2, 2)
plt.hist(poisson_sample, bins=range(0, max(poisson_sample)+1), density=True, color='lightgreen', alpha=0.7, label='Poisson')
plt.title('Poisson Distribution (Mean={:.2f}, Variance={:.2f})'.format(poisson_mean, poisson_variance))
plt.xlabel('Number of Events')
plt.ylabel('Probability')
plt.legend()
plt.grid(True)

plt.tight_layout()
plt.show()
```
